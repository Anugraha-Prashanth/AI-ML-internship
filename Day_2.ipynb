```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
#scikitlearn
#Load the dataset
file_path='/titanic uncleaned.csv'
titanic_data=pd.read_csv(file_path)

#select numerical columns for scaling
numerical_columns = ['Age', 'Fare', 'Pclass']

#Drop rows with missing values for simplicity
titanic_data.dropna(subset=numerical_columns,inplace=True)

#Initialize scalers
minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

#Apply Min-Max Scaling
titanic_data['Age_MinMax'] = minmax_scaler.fit_transform(titanic_data[['Age']])
titanic_data['Fare_MinMax'] = minmax_scaler.fit_transform(titanic_data[['Fare']])
#Apply Standard Scaling
titanic_data['Age_Standard'] = standard_scaler.fit_transform(titanic_data[['Age']])
titanic_data['Fare_Standard'] = standard_scaler.fit_transform(titanic_data[['Fare']])
```

```python
df['Age_mean_inputed']=df['Age'].fillna(df['Age'].mean())
df['Age_median_inputed']=df['Age'].fillna(df['Age'].median())

df['Department_mode_inputed']=df['Department'].fillna(df['Department'].mode())

df['Education_mode_inputed']=df['Department'].fillna(df['Department'].mode())

df['Monthly_Salary_mean_inputed']=df['Monthly_Salary'].fillna(df['Monthly_Salary'].mean())
df['Monthly_Salary_median_inputed']=df['Monthly_Salary'].fillna(df['Monthly_Salary'].median())

df['City_mode_inputed']=df['City'].fillna(df['City'].mode())
```
